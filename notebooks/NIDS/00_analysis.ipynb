### 0. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import numpy as np
import seaborn as sns
```

### 1. Data Overview

```python
df = pd.read_csv("WSN-DS.csv")
```

```python
attack_type_counts = df['Attack type'].value_counts()
attack_type_counts.plot(kind='bar', color=plt.cm.Paired(range(len(attack_type_counts))))
plt.title('Class Distribution')
plt.xlabel('Attack Type')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.show()
```

```python
encoder = LabelEncoder()
df['Attack_type_index'] = encoder.fit_transform(df['Attack type'])
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```
